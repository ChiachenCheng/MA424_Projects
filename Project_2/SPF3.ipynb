```r
# generating an exponetial distribution variable
exponetial = function(i){
    U = runif(1)
    return(-1.0/i*log(U))
}
```

```r
# generating a f(x) = 1/2(1+x)e^{-x} distribution variable
Sc_distribution = function(){
    repeat{
        Y = exponetial(0.5)
        U = runif(1)
        if (U <= (1+Y)/2*exp((1-Y)/2)){
            break
        }
    }
    return(Y)
}
```

```r
# generating a F(x) = 1- e^{-3x^2} distribution variable
Sw_distribution = function(){
    U = runif(1)
    return(sqrt(-1/3.0*log(U)))
}
```

```r
set.seed(1)
K = 10000
```

```r
# generating K S_C service times
Tsc = c()
k = 0
while (k < K){
    k = k + 1
    Tsc = c(Tsc, Sc_distribution())
}
```

```r
# generating K S_W service times
Tsw = c()
k = 0
while (k < K){
    k = k + 1
    Tsw = c(Tsw, Sw_distribution())
}
```

```r
x = c(0:1000)
x = x/100 # this is introduced to print density
n = 1 # the variable controlls the constants that densities multiply, see report
hist(x=Tsc, breaks=15*n)
lines(x, 0.5*(1+x)*exp(-x)*K/n) # f(x)*K/n
```

```r
hist(x=Tsw, breaks=15*n)
lines(x, 6*x*exp(-3*x*x)*K*0.1/n) # F'(x)*K/(10n)
```
